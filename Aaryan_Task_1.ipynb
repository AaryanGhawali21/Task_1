```python
import pandas as pd

df = pd.read_csv('marketing_campaign.csv', sep='\t')  # if it's tab-separated
df.head()
```

```python
df.isnull().sum()
df = df.dropna()  # or df.fillna(value)
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
df['Education'] = df['Education'].str.lower().str.strip()
df['Marital_Status'] = df['Marital_Status'].str.lower().str.replace('-', '_')
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.dtypes
```

```python
df['income'] = df['income'].astype(float)
```

```python
df.to_csv('marketing_campaign_cleaned.csv', index=False)
```
